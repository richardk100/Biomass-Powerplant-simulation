```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('biomass_power_plant_data.csv')
print(df.head())
print(df.columns)
```

```python
#Constants for conversion from MJ to kwh(1 kWh = 3.6 MJ)
MJ_to_kWh = 1/3.6

#Calculate the daily output in MJ
#Assuming the columns are named as 'Feedstock availability(tons)',
#'Calorific value(MJ/kg)', and 'Conversion Efficiency'
df['Daily Energy Output in MJ']=(
df['Feedstock Availability (tons)']* 1000*
df['Calorific Value (MJ/kg)']*
df['Conversion Efficiency'])
print(df.columns)
```

```python
#Convert the daily ouput to kwh
df['Daily Energy Ouput (kWh)'] = (
    df['Daily Energy Output in MJ'] * MJ_to_kWh)

#Plot the Daily Energy Output
plt.figure(figsize = (10, 6))
plt.plot(df['Day'], df['Daily Energy Ouput (kWh)'], 'o-')
plt.title('Daily Energy Output of a Biomass Powerplant')
plt.xlabel('Day')
plt.ylabel('Energy Output (kWh)')
plt.grid(True)
plt.show()
```
